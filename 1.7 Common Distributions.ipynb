```python
# Bernouli (2 possible outcomes, 1 or 0)
import sklearn
import numpy as np
from scipy.stats import bernoulli

data = sklearn.datasets.load_breast_cancer()
target_data = data['target']

p = np.sum(target_data) / len(target_data) # Not considering any features/factors

print(f"The probability of breast cancer in the dataset is {(p * 100).round(2)}%")

# Also use library
p = bernoulli.pmf(1, 0.6)
```

```python
# Poisson (Discrete, but outcomes are less likely as you go away from average)

import math
from scipy.stats import poisson


def p(x, theta):
    return (theta ** x * math.exp(-theta)) / math.factorial(x)

# If on avg you meet 4 new ppl a month, what is prob of meeting 20 new ppl?

p = p(20, 4)
print(p)

p = poisson.pmf(20, 4)
print(p)
```

```python
# Exponential Distribution (Continous: Less likely as you go forward from 0)
# p(x) = theta * e ^ -theta * x - Must integrate to find prob over an interval
# theta = 1/E[x]

from scipy.integrate import quad

# Example: Google page loading time average = 0.5 secs, prob over 2 secs

def p(x, theta):
    return theta * math.exp(-theta * x)

def g(x):
    return p(x, 2)

print(f"P(Wait Time > 2 secs) given avg of 0.5 is {round((quad(g, 2, 100))[0] * 100, 2)}%")
```

```python
# Normal/Gaussian Distribution (Continous: Skew = 0, Kurt = 3)

from scipy.stats import norm

mean = 4.1
std = 0.3

print(f"The prob of a GPA greater than 4 is {norm.sf(4, loc=mean, scale=std):.2f}")
print(f"The prob of a GPA Less than 3.5 is {norm.cdf(3.5, loc=mean, scale=std):.2f}")


```
