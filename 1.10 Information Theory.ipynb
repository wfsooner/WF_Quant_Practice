```python
"""Information Theory

Highly unlikely events give us much more information than likely ones

We define this as Information being inversely related to probability

I = log(1/p(x))
"""
```

```python
from numpy import log
import numpy as np
import math
```

```python
""" Chaos/Entropy

Chaos is measured by entropy, meant to evaluate outcomes that have not happened yet but have p(x)
Chaos/Entropy measures the uncertainty of an event

For below assume 4 events with distint probabilities
"""
```

```python
p_x = np.array([0.7, 0.2, 0.099, 0.001])
I_x = np.log2(p_x)

E_I_x = np.mean(p_x * I_x) # Average information

print(E_I_x)
```

```python
def calculate_entropy(p_x):
    entropy = 0
    for p in p_x:
        entropy -= p * math.log(p)
    return entropy

p_x = [0.4, 0.2, 0.2, 0.2]

entropy = calculate_entropy(p_x)
print(entropy)

"""
With a continous distribution:

entropy = int( p(x) * log(1/p(x)) )

"""
```

```python
""" Entropy, Normalized Entropy, Extra Entropy, KL Divergence

Normalized entropy = entropy / maximum entropy 

maximum entropy is when all p(x) are the same

H = int( p(x) * log(1/p(x)) )

Cross entropy = int( q(x) * log(1/p(x)) )
The more similar q(x) and p(x) the smaller it is
The more different they are the large the cross entropy

Extra entropy = cross entropy - entropy 
Extra entropy for q(x) = H(q, p) - H(p)
Extra Entropy =/aka KL Divergence

Extra Entropy = sum( p(x) * log( p(x)/q(x) ) )

"""
```

```python
"""Mutual Information (MI)

MI measures the dependence between 2 distributions

MI = H(p(x, y) , p(x)p(y))
MI = the KL divergence between p(x n y) and p(x) * p(y)
MI = sum( sum( p(x, y) * log( p(x, y)/p(x)p(y) ) ) ) sums for x and y

MI = 0 means independent

"""
```

```python
# Practice
p_x_y = np.array([[0.1, 0.2], # x = 0 is first column, x = 1 is second column
                  [0.4, 0.1], # y = 0 is first row, y = 1 is second row, y = 2 is third row
                  [0.1, 0.1]])

p_x = np.sum(p_x_y, axis=0)
p_y = np.sum(p_x_y, axis=1)

entropy_x = np.sum(p_x * log(1/p_x)) 

entropy_y = np.sum(p_y * log(1/p_y))

print("Entropy of X: ", entropy_x)
print("Entropy of Y: ", entropy_y)

# Cross Entropy of p(y|x=0) and p(y|x=1)
p_x0 = np.sum(p_x_y[:, 0])  # marginal probability of x=0
p_x1 = np.sum(p_x_y[:, 1])  # marginal probability of x=1

p_y_given_x0 = p_x_y[:, 0] / p_x0  # conditional distribution of y given x=0
p_y_given_x1 = p_x_y[:, 1] / p_x1  # conditional distribution of y given x=1

cross_entropy = np.sum(p_y_given_x0 * np.log2(1 / p_y_given_x1))

print("Cross Entropy: ", cross_entropy)

kl_divergence = cross_entropy - np.sum(p_y_given_x1 * np.log2(1 / p_y_given_x1))

print("KL Divergence: ", kl_divergence)

entropy_y_x0 = np.sum(p_y_given_x0 * np.log2(1 / p_y_given_x0))

print("Entropy of Y given X=0: ", entropy_y_x0)

px_py = p_y.reshape(-1, 1) @ p_x.reshape(1, -1)

print("P(X,Y): \n", px_py)

MI_xy = np.sum(p_x_y * np.log2(p_x_y / px_py))

print("MI(X,Y): ", MI_xy)
```

```python
from scipy import stats, integrate

# Define two continuous distributions
# p(x) = Normal(mu=0, sigma=1) — "true" distribution
# q(x) = Normal(mu=1, sigma=2) — approximate distribution
p = stats.norm(loc=0, scale=1)
q = stats.norm(loc=1, scale=2)

x = np.linspace(-10, 10, 10000)  # grid of x values for numerical integration
dx = x[1] - x[0]                 # step size for Riemann sum approximation

p_x = p.pdf(x)  # evaluate p(x) at each point
q_x = q.pdf(x)  # evaluate q(x) at each point

# Entropy: H(p) = integral( p(x) * log(1/p(x)) ) — measures uncertainty in p
entropy_p = np.sum(p_x * np.log(1 / np.where(p_x > 0, p_x, 1)) * dx)

# Entropy: H(q) = integral( q(x) * log(1/q(x)) ) — measures uncertainty in q
entropy_q = np.sum(q_x * np.log(1 / np.where(q_x > 0, q_x, 1)) * dx)

# Cross Entropy: H(q, p) = integral( q(x) * log(1/p(x)) )
# How many bits needed to encode q using p's encoding scheme
cross_entropy_qp = np.sum(q_x * np.log(1 / np.where(p_x > 0, p_x, 1)) * dx)

# Cross Entropy: H(p, q) = integral( p(x) * log(1/q(x)) )
cross_entropy_pq = np.sum(p_x * np.log(1 / np.where(q_x > 0, q_x, 1)) * dx)

# Extra Entropy (KL Divergence): D(q||p) = H(q, p) - H(q) — extra bits from using p instead of q
kl_q_given_p = cross_entropy_qp - entropy_q

# Extra Entropy (KL Divergence): D(p||q) = H(p, q) - H(p) — note: D(p||q) != D(q||p)
kl_p_given_q = cross_entropy_pq - entropy_p

# Mutual Information: MI = integral( p(x,y) * log( p(x,y) / p(x)*p(y) ) )
# For two independent 1D distributions, MI = 0 since p(x,y) = p(x)*p(y)
# To make it interesting, we compute MI between X~p and Y = X + noise
np.random.seed(42)
samples_x = p.rvs(10000)                          # draw samples from p(x)
samples_y = samples_x + np.random.normal(0, 0.5, 10000)  # Y is correlated with X

# Estimate MI using binned joint histogram
hist_2d, x_edges, y_edges = np.histogram2d(samples_x, samples_y, bins=50, density=True)
dx_bin = x_edges[1] - x_edges[0]
dy_bin = y_edges[1] - y_edges[0]

p_xy = hist_2d * dx_bin * dy_bin                  # joint distribution p(x, y)
p_x_marg = p_xy.sum(axis=1, keepdims=True)        # marginal p(x)
p_y_marg = p_xy.sum(axis=0, keepdims=True)        # marginal p(y)
p_x_p_y = p_x_marg @ p_y_marg                     # product of marginals p(x)*p(y)

mask = (p_xy > 0) & (p_x_p_y > 0)                # avoid log(0)
MI = np.sum(p_xy[mask] * np.log(p_xy[mask] / p_x_p_y[mask]))  # KL divergence between joint and product

print(f"Entropy H(p):          {entropy_p:.4f}")
print(f"Entropy H(q):          {entropy_q:.4f}")
print(f"Cross Entropy H(q,p):  {cross_entropy_qp:.4f}")
print(f"Cross Entropy H(p,q):  {cross_entropy_pq:.4f}")
print(f"KL D(q||p):            {kl_q_given_p:.4f}  (extra entropy encoding q with p)")
print(f"KL D(p||q):            {kl_p_given_q:.4f}  (extra entropy encoding p with q)")
print(f"MI(X, Y):              {MI:.4f}  (0 = independent, higher = more dependent)")
```
