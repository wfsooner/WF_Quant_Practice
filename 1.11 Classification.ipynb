```python
""" Logistic Regression

Binary Classification:
    Sigmoid Function: sigma(x) = 1/(1+e^-x)
    This function models the probability of 1 happening
    We can assume success is 1 and failure is 1 - success

The goal is to find the weight w such that it gives us the probability of 1 and 0 given an input x
We have a function q, if q given w is > 0.5 we say 1 is more likely
Vice versa for 0
We devise this function q so that it is an error function
We want to minimize the error of our predictions

The error of our q uses cross entropy
Hhat/Average Error = 1/n * Sum(cross entropy for each sample)

"""
```

```python
import numpy as np
from numpy import log2
from numpy import log
import math
```

```python
# Practice Calculating Average Cross Entropy

from logging import exception


q = np.array([[0.2, 0.8], [0.1, 0.9], [0.9, 0.1]])
x = np.array([[0, 1], [0, 1], [1, 0]])

if len(q) != len(x):
    raise exception

n = len(q)
cross_entropy_list = []
for i in range(n):
    q_i = q[i]
    x_i = x[i]

    pred_loc = np.argmax(x_i)
    cross_entropy = x_i[pred_loc] * log2(1/q_i[pred_loc])
    cross_entropy_list.append(cross_entropy)


average_cross_entropy = np.array(cross_entropy_list).mean()
print(f"Avg Cross Entropy: {average_cross_entropy}")

"""
Fast way to do it:

true_indices = np.argmax(x, axis=1)                    # index of true class for each sample
q_true = q[np.arange(n), true_indices]                 # predicted prob for true class
average_cross_entropy = np.mean(log2(1 / q_true))
"""


```

```python
# Read and load the data, scale the features

from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
import pandas as pd
from sklearn.preprocessing import StandardScaler


data = load_breast_cancer() # Load in data from sklearn

# Put data into dataframes
X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target

# Preprocces features
scaler = StandardScaler()
scaled_X = pd.DataFrame(scaler.fit_transform(X), columns=data.feature_names)
```

```python
# Try standard Logistic Regression

clf = LogisticRegression(random_state=0).fit(scaled_X, y)
predictions = clf.predict(scaled_X)

```
