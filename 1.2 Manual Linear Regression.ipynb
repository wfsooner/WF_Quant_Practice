```python
"""
The goal of this file is to perform minimization of the sum of squared errors between a linear function and a set of data points.
The steps are as follows:
1. identify and define the x and y values of the data points
2. Define the linear function I will be using. This function should have phi1, phi2... and w1, w2... as weights of the phis
3. I need to define the derivative of the function with respect to the weights.
4. I need to define the sum of squared errors function
5. I need to perform gradient descent to find the weights which give the lowest error
6. I need to plot the original data points and the fitted line on the same plot


"""
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 2, 3])
y = np.array([1, 1, 2, 2])
```

```python
def f(x, w):
    w1 = w[0]
    w2 = w[1]
    return w1*x + w2

def sum_of_squared_errors(x, y, w):
    return np.sum((y - f(x, w))**2) * (1/len(x))

def f_prime_w(x):
    return np.array([x, 1])

```

```python
def min_f(x, y, initial_w, eta, n):
    w = initial_w
    loss_history = []  # to track progress
    
    for i in range(n):
        # Step 1: Compute gradient of loss function
        # The gradient is the sum over all data points of: (residual) * (derivative of f w.r.t. w)
        # Start with zero gradient
        gradient = np.array([0.0, 0.0])
        
        # Loop through each data point
        for j in range(len(x)):
            # Calculate prediction for this point
            prediction = f(x[j], w)
            
            # Calculate residual (error) = prediction - actual
            # This is the chain rule: derivative of (y - f)^2 gives us (f - y)
            residual = prediction - y[j]
            
            # Get the derivative of f with respect to w for this x value
            # This is [x, 1] which represents [∂f/∂w1, ∂f/∂w2]
            df_dw = f_prime_w(x[j])
            
            # Accumulate: gradient += residual * df_dw
            # This is the chain rule: d/dw[(f - y)^2] = 2*(f - y) * df/dw
            # (the 2 is handled by not having 0.5 in our loss, or gets absorbed into learning rate)
            gradient = gradient + residual * df_dw
        
        # Average the gradient over all data points (to match our MSE definition)
        gradient = gradient * (1/len(x))
        
        # Step 2: Update weights using gradient descent
        w = w - eta * gradient
        
        # Step 3: Track the loss
        current_loss = sum_of_squared_errors(x, y, w)
        loss_history.append(current_loss)
    
    return w, loss_history


min_f(x, y, np.array([1.0, 1.0]), 0.1, 800)
```

```python
# Data for multivariable regression
a = np.array([1, 2, 3, 4, 5])
b = np.array([2, 3, 1, 4, 3])
z = np.array([5, 8, 7, 12, 11])
```

```python
def g(a, b, w):
    w1 = w[0]
    w2 = w[1]
    w3 = w[2]
    
    return w1*a + w2*b + w3

def g_prime_w(a, b):
    return np.array([a, b, 1])

def sum_of_squared_errors(a, b, z, w):
    return np.sum((z - g(a, b, w))**2) * (1/len(a))
```

```python
def min_g(a, b, z, initial_w, eta, n):
    w = initial_w

    if len(a) != len(b) and len(a) != len(z):
        quit

    error_history = []

    for i in range(n):
        gradient = np.array([0.0, 0.0, 0.0])

        predictions = g(a, b, w)  # Compute all 5 predictions at once: shape (5,)
        residuals = predictions - z  # Compute all 5 residuals at once: shape (5,)

        # Vectorized gradient calculation - replaces inner loop!
        # residuals @ a: dot product sums up (residual_i * a_i) for all i → scalar
        # residuals @ b: dot product sums up (residual_i * b_i) for all i → scalar  
        # np.sum(residuals): sums all residuals (equivalent to residuals @ ones) → scalar
        gradient = gradient + [residuals @ a, residuals @ b, np.sum(residuals)]

        gradient /= len(a)  # Average gradient over all data points

        w = w - eta * gradient  # Gradient descent step

        error = sum_of_squared_errors(a, b, z, w)  # Track loss for this iteration
        error_history.append(error)

    final_error = sum_of_squared_errors(a, b, z, w)

    print(f"Weights {w} gets an error {final_error}")

    return w, error_history


min_g(a, b, z, np.array([1.0, 1.0, 1.0]), 0.1, 1000)
print('')
```
