```python
from scipy.integrate import quad
import numpy as np
import matplotlib.pyplot as plt
```

```python
def p(x):
    return 0.5*x

result, error = quad(p, 1, 2) # using the pdf p(x) to find the probability of a <= x <= b
result
```

```python
# Histogram practice
from matplotlib.ticker import PercentFormatter


x = np.array([1, 4, 5, 3,2, 2,2, 1,1])

n, bins, patches = plt.hist(x, 20, facecolor='blue', alpha=0.7, weights=np.ones(len(x)) / len(x))
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
plt.show()
```

```python
# cdf practice
from scipy.stats import norm

a = norm.cdf(-1, loc=0, scale=1)
a
```

```python
# How to get E[X]

from scipy.integrate import quad

def e(x):
    #                  p(x)                x
    return (x ** 3 - x**2 - 0.1*x + 0.6) * x

quad(e, 0, 1.54337) # This is E[X] for the function p(x) * x defined in def e(X):
```
